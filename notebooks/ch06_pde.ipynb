   ## 6.1 放物型偏微分方程式

```julia
N=20

function main()

    u = Array{Float64}(undef, N+1)
    w = Array{Float64}(undef, N+1)
    k = 0.001
    h = 1.0 / N
    r = k / h^2
    s = 1.0 - 2.0 * r
    
    for i in 1:N+1
       w[i] = 0.0
    end
    for i in 2:N
       u[i] = 1.0
    end
    u[1] = 0.0
    u[N+1]=0.0

    for j in 2:201

        if j % 10 == 1
            @printf("%5.3lf ", j * k)
            for i in 1:2:N+1
                @printf("%5.3lf ", u[i])
            end
            @printf("\n")
        end

        for i in 2:N
            w[i] = r * (u[i+1] + u[i-1]) + s * u[i]
        end

        for i in 1:N+1
            u[i] = w[i]
        end
        
    end

    return 0
        
end
    
main()
```
